Dictionary

Dictionary dengan Python adalah kumpulan pasangan key-value yang tidak berurutan. Dictionary dapat digunakan untuk menyimpan data berukuran kecil hingga besar. Untuk mengakses data, kita harus mengetahui kuncinya (key). Di Python, dictionary didefinisikan dengan kurung kurawal dan definisi tambahan berikut:
• Setiap elemen pasangan key-value dipisahkan dengan koma (,)
• Key dan Value dipisahkan dengan titik dua (:)
• Key dan Value dapat berupa variabel/objek jenis apa pun

```python
a = {"nama": "Nirmala", "Usia": 20}
```

```python
type(a)
```

```python
a["nama"]
```

```python
a["Nirmala"]
```

```python
a["Usia"]
```

Dictionary bukan termasuk dalam implementasi urutan (sequence), sehingga tidak bisa dipanggil dengan urutan indeks. Misalnya berikut dicoba dengan indeks 0, tetapi menghasilkan eror (KeyErros) karena tidak ada kunci (key) 0.

```python
a[0]
```
